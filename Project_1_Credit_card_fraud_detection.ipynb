Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
# loading the dataset to pandas DataFrame
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
#Printing 1st 5rows of the dataset
credit_card_data.head()
```

```python
#printing last 5 rows of the dataset
credit_card_data.tail()
```

```python
#Dataset information
credit_card_data.info()
```

```python
#checking the missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transaction(0) and fraudulent transaction(1)
credit_card_data['Class'].value_counts()
```

This data set is highly unbalanced


0->Normal Transaction
1->fraudulent Transaction

```python
#seperating the data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both Transactions
credit_card_data.groupby('Class').mean()
```

Under-Sampling

build a Sample dataset containing similar distribution of normal transaction and fraudulent transactions

Number of Fraudulent Transactions-->492

```python
legit_sample=legit.sample(n=492)
```

Concatenating two Dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0) #axis=0(adding row wise below the legit_sample values) axis=1(adding column wise)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and targets

 Storing Features in X (here time(v1,v2,....))

 Storing Targets in Y(here Class(0,1))

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

To Split the data into traing data and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2) #20% of data is for testing and remaining 80% of data for training
#Y in stratify means there will be equal distibution of classes in X_train and X_test
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

Logical Regression

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
#Training the logisticRegression model with Training Data
model =LogisticRegression()
model.fit(X_train,Y_train)
#x_train contains all the features of the Training data and Y_train contains the corresponding labels(0's and 1's)
```

model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data:',training_data_accuracy)
```

```python
#accuracy on testing data
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on testing data',testing_data_accuracy)
```
